```python
df=pd.read_excel("Customer Call List.xlsx")
df
#creating dataframe
```

```python
df.drop_duplicates()
#droping dublicates rows, there are 2 same rows
```

```python
df=df.drop(columns = "Not_Useful_Column")
df.head()
# We can drop every columns with this code.
```

```python
df["Last_Name"]=df["Last_Name"].str.strip('./*_-')
df["Last_Name"]=df["Last_Name"].str.strip('')
df.head()
#We can drop every character and blank at left and right sides of text
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^0-9]','',regex=True)
df.head()
#We can delete every character that we don't want in the "Phone_Number" column
```

```python
#df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',' , 2 , expand=True)
#df
#Some address have different info so we can split them by comma and creat news columns
```

```python
df["Paying Customer"]=df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"]=df["Paying Customer"].str.replace('No','N')
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('No','N')
df.head()
#I changed all "Yes,No" so all columns became same format. If we need, We could all of them "Yes,No"
```
